````python
import sys
sys.path.append("..")

from sicm import io
from sicm.sicm import Signal
import numpy as np
import os
from sicm.plots import plots
from sicm.mathops import mathops
from sicm.utils import utils

%load_ext autoreload
%autoreload 2
````

````python
datadir = r"S:\UsersData\Martin\2019\03_Mar\20\tip_current_2um"
exp_names = [str(np.int(x)) for x in np.arange(3, 12)]

X = []
Y = []
for name in exp_names:
    
    files, date = io.get_files(datadir, name)
    result_ = io.load_result(files, name)
    
    t = np.cumsum(result_["dt(s)"])
    y = result_["Current1(A)"] * 1e9
    
    t, y = mathops.rolling_mean(t, y, 10)
    sel = np.nonzero(y > 2)[0][100:-100]
    
    X.append(t[sel])
    Y.append(y[sel])

fname = datadir + "\\" + "".join(exp_names)
plots.plot_generic(X, Y, [r"$t\ [s]$"], [r"$I\ [nA]$"], alpha = 0.3, fname = fname)
````
